```python
import numpy as np
import pandas as pd
from IPython.display import display, HTML
```

```python
data = pd.read_csv("./Q1/popular_board_game.csv")
display(data)
```

```python
print(np.unique(data["min_players"]))
print(np.unique(data["playtime"]))
```

```python
data["support_solo"] = np.where(data['min_players'] == 1, True, False)
display(data)
```

```python
df = data
# Assuming you have a DataFrame named 'df' with the specified columns
# You can group the data by the 'support_solo' column and calculate summary metrics

# Group the data by 'support_solo' and 'category'
grouped = df.groupby(['support_solo', 'category'])

# Calculate the summary metrics for each group
summary_df = grouped.agg({
    'name': 'count',
    'avg_rating': 'mean',
    'num_ratings': 'max',
    'playtime_num': 'mean'
}).reset_index()

# Rename the columns
summary_df.columns = ['Group', 'Category', 'Number of Games', 'Average Rating', 'Highest Number of Ratings', 'Average Playtime']

# Round the 'Average Rating' and 'Average Playtime' columns to two decimal places
summary_df['Average Rating'] = summary_df['Average Rating'].round(2)
summary_df['Average Playtime'] = summary_df['Average Playtime'].round(2)

summary_df['Group'] = summary_df['Group'].map({True: 'Solo supported', False: 'Solo not supported'})

summary_df = summary_df.set_index(['Group', 'Category'])
display(summary_df)
```

```python
df = data
# Assuming you have a DataFrame named 'df' with the specified columns
# You can group the data by the 'support_solo' column and calculate summary metrics

# Group the data by 'support_solo' and 'category'
grouped = df.groupby(['support_solo', 'category'])

# Calculate the summary metrics for each group
summary_df = grouped.agg({
    'name': 'count',
    'avg_rating': 'mean',
    'num_ratings': 'max',
    'playtime_num': 'mean'
}).reset_index()

# Create a new DataFrame for the 'Highest Number of Ratings' information
max_ratings_idx = grouped['num_ratings'].idxmax()
max_ratings_games = df.loc[max_ratings_idx, ['support_solo', 'category', 'name']]
max_ratings_games.rename(columns={'name': 'Highest Number of Ratings'}, inplace=True)

# Merge the 'Highest Number of Ratings' information back into the summary DataFrame
summary_df = pd.merge(summary_df, max_ratings_games, on=['support_solo', 'category'])


display(summary_df)
# Rename the columns
summary_df.columns = ['Group', 'Category', 'Number of Games', 'Average Rating', 'Number of Ratings', 'Average Playtime', 'Highest Number of Ratings']
column_to_drop = 'Number of Ratings'
summary_df = summary_df.drop(column_to_drop, axis=1)
# Round the 'Average Rating' and 'Average Playtime' columns to two decimal places
summary_df['Average Rating'] = summary_df['Average Rating'].round(2)
summary_df['Average Playtime'] = summary_df['Average Playtime'].round(2)

summary_df['Group'] = summary_df['Group'].map({True: 'Solo supported', False: 'Solo not supported'})

summary_df = summary_df.set_index(['Group', 'Category'])
display(summary_df)
```
